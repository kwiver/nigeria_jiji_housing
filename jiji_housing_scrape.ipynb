```python
# import libraries
import requests
import re
import pandas as pd
```

```python
def get_attrs_value(attrs, key_name):
    for attr in attrs:
        if attr.get("name", "").lower() == key_name.lower():
            return str(attr.get("value", "")).strip()
    return None
```

```python
def extract_furnishing(attrs):
    for attr in attrs:
        if attr.get("name") == "Furnishing":
            value = attr.get("value", "")
            furnishing = value.lower()
            if "semi" in furnishing:
                return "semi-furnished"
            elif "unfurnished" in furnishing:
                return "unfurnished"
            elif "furnished" in furnishing:
                return "furnished"
             
    return None
```

```python
def fetch_json(page):
    url = "https://jiji.ng/api_web/v1/listing"
    params = {
        "slug": "houses-apartments-for-sale",
        "page": 1,
        "webp": "true"
    }
    headers = {
        "User-Agent": "Mozilla/5.0"
    }
    try:
        response = requests.get(url, params=params)
        response.raise_for_status()
        data = response.json()
    except requests.RequestException:
        print(f"[Page] {page} . Request error")
        return []
    except ValueError:
        print(f"[Page] {page} . Decode error")
        
    adverts = data.get("adverts_list", {}).get("adverts", [])
    if not isinstance(adverts, list):
        print(f"An error occured. Expected list but got {type(adverts)}")
        return[]
    return adverts
```

```python
def main():
    all_ads = []
    
    for page in range(1, 101):
        ads = fetch_json(page)
        print(f"page {page}: {len(ads)} Found")

        for ad in ads:
            if isinstance(ad, dict):
                attrs = ad.get("attrs", [])
                title = ad.get("title", "")
                region = ad.get("region", "")
                region_name = ad.get("region_name", "")
                region_parent_name = ad.get("region_parent_name", "")
                price_title = ad.get("price_title", "")
                property_size_ = get_attrs_value(attrs, "Property size")
                property_size = property_size_
                bedrooms_ = get_attrs_value(attrs, "Bedrooms")
                bedrooms = bedrooms_
                bathrooms_ = get_attrs_value(attrs, "Bathrooms")
                bathrooms = bathrooms_
                furnishing = extract_furnishing(attrs)
                package_type = extract_package_type(ad)

                if price_title:
                    all_ads.append({
                        "title": title,
                        "region": region,
                        "region_name": region_name,
                        "region_parent_name": region_parent_name,
                        "price_title": price_title,
                        "property_size": property_size,
                        "bedrooms": bedrooms,
                        "bathrooms": bathrooms,
                        "furnishing": furnishing,
                        "is_boost": package_type
                    })
    if all_ads:
        df = pd.DataFrame(all_ads)
        df.to_csv("jiji_housing_raw.csv", index=False)
        print("successfully extract data from jiji")
    else:
        print("Nothing to extract")
```

```python
if __name__ == "__main__":
    main()
```
